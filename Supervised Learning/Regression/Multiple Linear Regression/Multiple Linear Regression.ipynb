# Multiple Linear Regression


### Import Libraries



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Load and read dataset

```python
weather_data_m = pd.read_csv("WeatherDataM.csv") 
# Display the data in the notebook
weather_data_m
weather_data_m.columns
```

```python
# Set the features of our model, these are our potential inputs
weather_features = ['Temperature (C)', 'Wind Speed (km/h)', 'Pressure (millibars)']

# Set the variable X to be all our input columns: Temperature, Wind Speed and Pressure
X = weather_data_m[weather_features]

# set y to be our output column: Humidity
y = weather_data_m.Humidity
```

```python
# plt.subplot enables us to plot mutliple graphs
# we produce scatter plots for Humidity against each of our input variables

plt.subplot(2,2,1)
plt.scatter(X['Temperature (C)'],y)
plt.subplot(2,2,2)
plt.scatter(X['Wind Speed (km/h)'],y)
plt.subplot(2,2,3)
plt.scatter(X['Pressure (millibars)'],y)
```

```python
from sklearn.linear_model import LinearRegression

mlr_model = LinearRegression()
mlr_model.fit(X, y)
```

```python
# Import library to produce a 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x1 = X["Temperature (C)"]
x2 = X["Wind Speed (km/h)"]

ax.scatter(x1, x2, y, c='r', marker='o')

# Set axis labels
ax.set_xlabel('Temperature (C)')
ax.set_ylabel('Wind Speed (km/h)')
ax.set_zlabel('Humidity')
```

```python
theta0 = mlr_model.intercept_
theta1, theta2, theta3 = mlr_model.coef_

theta0, theta1, theta2, theta3
```

```python
y_pred = mlr_model.predict([[15, 21, 1015]])
print(y_pred)
```

### Conclusion

This analysis demonstrates the application of **Multiple Linear Regression (MLR)** to model the relationship between several weather-related features — *Temperature (C), Wind Speed (km/h), and Pressure (millibars)* — and the target variable *Humidity*. The dataset was loaded and visualized through scatter plots and a 3D plot to inspect how each feature relates to the target.

The MLR model was trained successfully, learning coefficients that describe how each input feature contributes to the prediction of Humidity. The model equation takes the form:

$$
\hat{y} = \theta_0 + \theta_1 (\text{Temperature}) + \theta_2 (\text{Wind Speed}) + \theta_3 (\text{Pressure})
$$

Predictions can be made by providing values for all three input features, allowing estimation of Humidity under varying weather conditions.

This example highlights the power and simplicity of linear models for predicting continuous outcomes when multiple features contribute linearly to the target. It also underscores the importance of providing the correct feature dimensions during prediction and interpreting model coefficients to understand feature impact.

Multiple Linear Regression proves to be a useful approach in weather forecasting applications where relationships between variables are expected to follow approximately linear patterns.
